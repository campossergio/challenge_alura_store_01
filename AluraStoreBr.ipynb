### Importação dos dados



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.head()
```

```python
# Usando o DataFrame loja1: calcular o faturamento total da loja 1

# Calculo do faturamento da Loja 1
total_faturamento = loja1['Preço'].sum()

# Mostrando resultado da Loja 1
print(f"O faturamento total da loja 1 é: R${total_faturamento:.2f}")
```

#1. Análise do faturamento


```python
# Calculando faturamento total de cada loja
faturamento_loja1 = loja1["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

# Exibindo os resultados
print("Faturamento total por loja:")
print(f"Faturamento total da Loja 1: R$ {faturamento_loja1:,.2f}")
print(f"Faturamento total da Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento total da Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento total da Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
# prompt: gerando gráficos barra e pizza destacando os pontos mais relevantes utilizando a biblioteca matplotlib

import matplotlib.pyplot as plt

# Dados de faturamento (substitua pelos seus dados reais)
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Gráfico de barras
plt.figure(figsize=(6, 4))
plt.bar(lojas, faturamentos, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Faturamento por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(faturamentos, labels=lojas, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Distribuição do Faturamento por Loja')
plt.axis('equal')  # Para garantir que o gráfico seja um círculo
plt.show()
```

# 2. Vendas por Categoria


```python
# prompt:  calculando a quantidade de produtos vendidos por categoria em cada loja.

# Agrupar por categoria e contar as vendas em cada loja
vendas_por_categoria_loja1 = loja1.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Preço'].count()
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Preço'].count()

# Exibir as categorias mais populares de cada loja
print("\nCategorias mais populares por loja:")
print("Loja 1:\n", vendas_por_categoria_loja1.sort_values(ascending=False))
print("\nLoja 2:\n", vendas_por_categoria_loja2.sort_values(ascending=False))
print("\nLoja 3:\n", vendas_por_categoria_loja3.sort_values(ascending=False))
print("\nLoja 4:\n", vendas_por_categoria_loja4.sort_values(ascending=False))
```

```python
# prompt: 'criando um gráfico de barra para exibir os dados da vendas por categoria separados por loja

# Criando um DataFrame com os dados de vendas por categoria e loja
vendas_por_categoria = pd.DataFrame({
    'Loja 1': vendas_por_categoria_loja1,
    'Loja 2': vendas_por_categoria_loja2,
    'Loja 3': vendas_por_categoria_loja3,
    'Loja 4': vendas_por_categoria_loja4
})

# Preencher valores ausentes com 0 (caso alguma categoria não exista em uma loja)
vendas_por_categoria = vendas_por_categoria.fillna(0)

# Criar o gráfico de barras
vendas_por_categoria.plot(kind='bar', figsize=(10, 6))
plt.title('Vendas por Categoria (por Loja)')
plt.xlabel('Categoria do Produto')
plt.ylabel('Número de Vendas')
plt.xticks(rotation=45)
plt.legend(title='Lojas')
plt.tight_layout()
plt.show()
```

```python
# prompt: criando um gráfico de barra e outro de pizza dos dados da venda por categoria

# Vendas por Categoria - Gráfico de Barras
plt.figure(figsize=(8, 6))
vendas_por_categoria_loja1.plot(kind='bar', color='skyblue', label='Loja 1')
vendas_por_categoria_loja2.plot(kind='bar', color='lightcoral', label='Loja 2')
vendas_por_categoria_loja3.plot(kind='bar', color='lightgreen', label='Loja 3')
vendas_por_categoria_loja4.plot(kind='bar', color='gold', label='Loja 4')
plt.title('Vendas por Categoria em cada Loja')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# Vendas por Categoria - Gráfico de Pizza Loja 1
plt.figure(figsize=(6, 6))
plt.pie(vendas_por_categoria_loja1, labels=vendas_por_categoria_loja1.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Vendas por Categoria (Loja 1)')
plt.axis('equal')
plt.show()

# Vendas por Categoria - Gráfico de Pizza Loja 2
plt.figure(figsize=(6, 6))
plt.pie(vendas_por_categoria_loja2, labels=vendas_por_categoria_loja2.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Vendas por Categoria (Loja 2)')
plt.axis('equal')
plt.show()

# Vendas por Categoria - Gráfico de Pizza Loja 3
plt.figure(figsize=(6, 6))
plt.pie(vendas_por_categoria_loja3, labels=vendas_por_categoria_loja3.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Vendas por Categoria (Loja 3)')
plt.axis('equal')
plt.show()

# Vendas por Categoria - Gráfico de Pizza Loja 4
plt.figure(figsize=(6, 6))
plt.pie(vendas_por_categoria_loja4, labels=vendas_por_categoria_loja4.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Vendas por Categoria (Loja 4)')
plt.axis('equal')
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# calcular a média das avaliações dos clientes para cada loja. O objetivo é entender a satisfação dos clientes com os produtos vendidos.

# Calcular a média das avaliações para cada loja
media_avaliacoes_loja1 = loja1['Avaliação da compra'].mean()
media_avaliacoes_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacoes_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacoes_loja4 = loja4['Avaliação da compra'].mean()

# Exibir as médias de avaliação
print("\nMédia de avaliações por loja:")
print(f"Loja 1: Média das Avaliações - Nota {media_avaliacoes_loja1:.2f}")
print(f"Loja 2: Média das Avaliações - Nota {media_avaliacoes_loja2:.2f}")
print(f"Loja 3: Média das Avaliações - Nota {media_avaliacoes_loja3:.2f}")
print(f"Loja 4: Média das Avaliações - Nota {media_avaliacoes_loja4:.2f}")
```

```python
# prompt: criando um gráfico de barras e outro de pizza para exibir a média das avaliações das lojas

# Dados de média de avaliações (substitua pelos seus dados reais)
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias_avaliacoes = [media_avaliacoes_loja1, media_avaliacoes_loja2, media_avaliacoes_loja3, media_avaliacoes_loja4]

# Gráfico de barras
plt.figure(figsize=(6, 4))
plt.bar(lojas, medias_avaliacoes, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Média de Avaliações por Loja')
plt.xlabel('Lojas')
plt.ylabel('Média das Avaliações')
plt.ylim(0, 5)  # Definir limite do eixo y de 0 a 5 (para avaliações)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(medias_avaliacoes, labels=lojas, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Distribuição da Média de Avaliações por Loja')
plt.axis('equal')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Criando uma função para analisar os produtos mais e menos vendidos
def analisar_produtos(loja, nome_loja):
    vendas_por_produto = loja.groupby('Produto')['Preço'].count().sort_values(ascending=False)
    mais_vendidos = vendas_por_produto.head(5)
    menos_vendidos = vendas_por_produto.tail(5)
    print(f"\nProdutos mais vendidos na {nome_loja}:")
    print(mais_vendidos)
    print(f"\nProdutos menos vendidos na {nome_loja}:")
    print(menos_vendidos)

# Analisar os produtos de cada loja
analisar_produtos(loja1, "Loja 1")
analisar_produtos(loja2, "Loja 2")
analisar_produtos(loja3, "Loja 3")
analisar_produtos(loja4, "Loja 4")
```

```python
# prompt: criando um gráfico de barra e outro de pizza para exibir os dados dos produtos mais vendidos e menos vendidos separados por loja

import matplotlib.pyplot as plt

def plot_products(loja, nome_loja):
    vendas_por_produto = loja.groupby('Produto')['Preço'].count().sort_values(ascending=False)
    mais_vendidos = vendas_por_produto.head(5)
    menos_vendidos = vendas_por_produto.tail(5)

    # Gráfico de barras para produtos mais vendidos
    plt.figure(figsize=(6, 4))
    plt.bar(mais_vendidos.index, mais_vendidos.values, color='skyblue')
    plt.title(f'Produtos Mais Vendidos - {nome_loja}')
    plt.xlabel('Produtos')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Gráfico de pizza para produtos mais vendidos
    plt.figure(figsize=(6, 6))
    plt.pie(mais_vendidos.values, labels=mais_vendidos.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Produtos Mais Vendidos - {nome_loja}')
    plt.axis('equal')
    plt.show()


    # Gráfico de barras para produtos menos vendidos
    plt.figure(figsize=(6, 4))
    plt.bar(menos_vendidos.index, menos_vendidos.values, color='lightcoral')
    plt.title(f'Produtos Menos Vendidos - {nome_loja}')
    plt.xlabel('Produtos')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Gráfico de pizza para produtos menos vendidos
    plt.figure(figsize=(6, 6))
    plt.pie(menos_vendidos.values, labels=menos_vendidos.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Produtos Menos Vendidos - {nome_loja}')
    plt.axis('equal')
    plt.show()

# Chamar a função para cada loja
plot_products(loja1, "Loja 1")
plot_products(loja2, "Loja 2")
plot_products(loja3, "Loja 3")
plot_products(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
# Neste passo, vamos calcular o custo médio de frete para cada loja. O objetivo é entender quanto, em média, está sendo gasto com frete para cada uma das lojas.

# Calcular o frete médio para cada loja
frete_medio_loja1 = loja1['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Exibir os fretes médios
print("\nFrete médio por loja:")
print(f"Loja 1: R$ {frete_medio_loja1:.2f}")
print(f"Loja 2: R$ {frete_medio_loja2:.2f}")
print(f"Loja 3: R$ {frete_medio_loja3:.2f}")
print(f"Loja 4: R$ {frete_medio_loja4:.2f}")
```

```python
# prompt: criando um gráfico de barra e outro de pizza para mostrar o custo médio do frete de cada loja

# Dados de frete médio (substitua pelos seus dados reais)
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes_medios = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

# Gráfico de barras
plt.figure(figsize=(6, 4))
plt.bar(lojas, fretes_medios, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Custo Médio do Frete por Loja')
plt.xlabel('Lojas')
plt.ylabel('Custo Médio do Frete (R$)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(fretes_medios, labels=lojas, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Distribuição do Custo Médio do Frete por Loja')
plt.axis('equal')  # Para garantir que o gráfico seja um círculo
plt.show()
```

```python
# prompt: criando gráfico de dispersão relacionando as informações de análise de faturamento, vendas por categoria, média de avaliações das lojas, produtos mais vendidos e menos vendidos e frete médio por lojas

import matplotlib.pyplot as plt

# Dados para o gráfico de dispersão (Exemplo - Substitua pelos seus dados)
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
avaliacoes = [media_avaliacoes_loja1, media_avaliacoes_loja2, media_avaliacoes_loja3, media_avaliacoes_loja4]
frete_medio = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))

# Adicionando as informações no gráfico
plt.scatter(faturamentos, avaliacoes, s=100, c=frete_medio, cmap='viridis', alpha=0.7) #s - tamanho do ponto. c - cor do ponto. alpha - transparencia do ponto

# Adicionando rótulos e título
plt.title('Relação entre Faturamento, Avaliações e Frete Médio')
plt.xlabel('Faturamento (R$)')
plt.ylabel('Média de Avaliações')

# Adicionando a barra de cores (Colorbar)
cbar = plt.colorbar()
cbar.set_label('Frete Médio (R$)')

# Adicionando os nomes das lojas como rótulos dos pontos
for i, loja in enumerate(lojas):
    plt.annotate(loja, (faturamentos[i], avaliacoes[i]), textcoords="offset points", xytext=(5,5), ha='left')

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
